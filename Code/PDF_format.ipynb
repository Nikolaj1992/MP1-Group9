# ETL on PDF

## Imports

```python
import os
import PyPDF2
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## Data Loading 

```python
def load_all_pdfs_pypdf2(folder_path):
    all_data = []

    for filename in os.listdir(folder_path):
        if filename.lower().endswith(".pdf"):
            file_path = os.path.join(folder_path, filename)
            try:
                with open(file_path, 'rb') as f:
                    reader = PyPDF2.PdfReader(f)
                    for page_num, page in enumerate(reader.pages):
                        text = page.extract_text()
                        if text:
                            lines = text.split('\n')
                            for line in lines:
                                clean_line = line.strip()
                                if clean_line:
                                    all_data.append({
                                        'file': filename,
                                        'page': page_num + 1,
                                        'line': clean_line
                                    })
            except Exception as e:
                print(f"Error reading {filename}: {e}")
    
    return pd.DataFrame(all_data)
```

## Clean PDF Data

```python
def clean_pdf_df(df):
    # Strip leading and trailing whitespace
    df['line'] = df['line'].str.strip()
    
    # Remove empty rows
    df = df[df['line'] != '']
    
    # Remove duplicate words in each 'line'
    df['line'] = df['line'].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))
    
    # Remove duplicate rows based on the 'line' column
    df = df.drop_duplicates(subset=['line'])
    
    # Reset the index
    return df.reset_index(drop=True)
```

## Explore or Transform

```python
def word_frequency(df):
    words = ' '.join(df['line']).split()
    return Counter(words)
```

## Example Usage

```python
print(os.getcwd())  # Check your working directory

folder_path = "../Data/pdfs"  # Your folder path

# Load and clean PDFs
raw_df = load_all_pdfs_pypdf2(folder_path)
cleaned_df = clean_pdf_df(raw_df)

# Word count (assuming you have this function defined)
word_count = word_frequency(cleaned_df)
print(word_count.most_common(5))  # Top 5 most common words

# Preview data
print(cleaned_df.head())
print(cleaned_df.tail())

# Save to CSV
cleaned_df.to_csv('../Data/cleaned_pdf_data.csv', index=False)

# Convert to DataFrame
top_words = word_count.most_common(5)
top_words_df = pd.DataFrame(top_words, columns=['Word', 'Frequency'])

# Save to CSV
top_words_df.to_csv('../Data/top_5_words.csv', index=False, sep=';', encoding='utf-8')
```

## Visualize PDF

```python
def plot_wordcloud(text):
    wc = WordCloud(width=1200, height=800, background_color='white').generate(text)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title("Word Cloud from PDFs")
    plt.show()

all_text = ' '.join(cleaned_df['line'])
plot_wordcloud(all_text)
```

```python
def plot_wordcloud(text, mask_image_path):
    # Load the Darth Vader image as the mask
    darth_mask = np.array(Image.open(mask_image_path))

    wc = WordCloud(width=1200, height=800, background_color='white', mask=darth_mask, contour_width=1, contour_color='black').generate(text)
    
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title("Word Cloud in Darth Vader Shape")
    plt.show()

all_text = ' '.join(cleaned_df['line'])
darth_image_path = '../Data/Vader.png'  # Replace this with the actual path to your unicorn image

# Call the function 
plot_wordcloud(all_text, darth_image_path)
```
